1. Approach 1: Triple loop brute force, O(n^3), O(1)

    import java.util.*;
    // T.C: O(n^3)
    // S.C: O(1)
    public class Test {
        public static void main(String[] args) {
            Scanner sc = new Scanner(System.in);
            int n = sc.nextInt();

            int[] arr = new int[n];
            for (int i = 0; i < n; i++) {
                arr[i] = sc.nextInt();
            }

            int maxLen = 0;

            for (int i = 0; i < n; i++) {
                for (int j = i; j < n; j++) {
                    int zeroCount = 0;
                    int oneCount = 0;

                    for (int k = i; k <= j; k++) {
                        if (arr[k] == 0) zeroCount++;
                        else oneCount++;
                    }

                    if (zeroCount == oneCount) {
                        int len = j - i + 1;
                        if (len > maxLen) maxLen = len;
                    }
                }
            }

            System.out.println(maxLen);
            sc.close();
        }
    }


2. Approach 2: Double loop with running balance, O(n^2), O(1)

    import java.util.*;
    // T.C: O(n^2)
    // S.C: O(1)
    public class Test {
        public static void main(String[] args) {

            int n = sc.nextInt();

            int[] arr = new int[n];
            for (int i = 0; i < n; i++) {
                arr[i] = sc.nextInt();
            }

            int maxLen = 0;

            for (int i = 0; i < n; i++) {
                int balance = 0; // +1 for 1, -1 for 0
                for (int j = i; j < n; j++) {
                    if (arr[j] == 1) balance++;
                    else balance--;

                    if (balance == 0) {
                        int len = j - i + 1;
                        if (len > maxLen) maxLen = len;
                    }
                }
            }

            System.out.println(maxLen);
            sc.close();
        }
    }


3. Approach 3: Prefix balance + HashMap, O(n) average, O(n)

    import java.util.*;
    // T.C: O(n) average
    // S.C: O(n)
    public class Test {
        public static void main(String[] args) {
            Scanner sc = new Scanner(System.in);
            int n = sc.nextInt();

            int[] arr = new int[n];
            for (int i = 0; i < n; i++) {
                arr[i] = sc.nextInt();
            }

            HashMap<Integer, Integer> firstIndex = new HashMap<>();
            firstIndex.put(0, -1); // balance 0 before array starts

            int balance = 0;
            int maxLen = 0;

            for (int i = 0; i < n; i++) {
                if (arr[i] == 1) balance++;
                else balance--;

                if (firstIndex.containsKey(balance)) {
                    int len = i - firstIndex.get(balance);
                    if (len > maxLen) maxLen = len;
                } else {
                    firstIndex.put(balance, i);
                }
            }

            System.out.println(maxLen);
            sc.close();
        }
    }


4. Approach 4: Prefix balance + direct array index, O(n), O(n)

    import java.util.*;
    // T.C: O(n)
    // S.C: O(n)
    public class Test {
        public static void main(String[] args) {
            Scanner sc = new Scanner(System.in);
            int n = sc.nextInt();

            int[] arr = new int[n];
            for (int i = 0; i < n; i++) {
                arr[i] = sc.nextInt();
            }

            int[] first = new int[2 * n + 1];
            Arrays.fill(first, Integer.MIN_VALUE);

            int offset = n;
            int balance = 0;
            int maxLen = 0;

            first[offset] = -1; // balance 0 seen at index -1

            for (int i = 0; i < n; i++) {
                if (arr[i] == 1) balance++;
                else balance--;

                int idx = balance + offset;

                if (first[idx] != Integer.MIN_VALUE) {
                    int len = i - first[idx];
                    if (len > maxLen) maxLen = len;
                } else {
                    first[idx] = i;
                }
            }

            System.out.println(maxLen);
            sc.close();
        }
    }


5. Approach 5: Prefix balances + sorting, O(n log n), O(n)

    import java.util.*;
    // T.C: O(n log n)
    // S.C: O(n)
    public class Test {
        public static void main(String[] args) {
            Scanner sc = new Scanner(System.in);
            int n = sc.nextInt();

            int[] arr = new int[n];
            for (int i = 0; i < n; i++) {
                arr[i] = sc.nextInt();
            }

            // pair: [balance, index]
            int[][] pref = new int[n + 1][2];
            pref[0][0] = 0;
            pref[0][1] = -1;

            int balance = 0;
            for (int i = 1; i <= n; i++) {
                if (arr[i - 1] == 1) balance++;
                else balance--;
                pref[i][0] = balance;
                pref[i][1] = i - 1;
            }

            Arrays.sort(pref, (a, b) -> {
                if (a[0] != b[0]) return Integer.compare(a[0], b[0]);
                return Integer.compare(a[1], b[1]);
            });

            int maxLen = 0;
            int i = 0;

            while (i <= n) {
                int j = i;
                int minIdx = pref[i][1];
                int maxIdx = pref[i][1];

                while (j <= n && pref[j][0] == pref[i][0]) {
                    if (pref[j][1] < minIdx) minIdx = pref[j][1];
                    if (pref[j][1] > maxIdx) maxIdx = pref[j][1];
                    j++;
                }

                int len = maxIdx - minIdx;
                if (len > maxLen) maxLen = len;

                i = j;
            }

            System.out.println(maxLen);
            sc.close();
        }
    }
